```python
add = np.zeros((len(train_df),18),dtype=np.float32)
stock = None
molecule_name_atom = None
for i,row in tqdm(train_df.iterrows(),total=len(train_df)):
    if stock != row[1]:
        stock = row[1]
        molecule_name_atom = magnetic_shielding_df[magnetic_shielding_df['molecule_name']==stock].values
    add[i,0] = molecule_name_atom[np.where(molecule_name_atom[:,1]==row[2])[0],2]
    add[i,1] = molecule_name_atom[np.where(molecule_name_atom[:,1]==row[3])[0],2]
    add[i,2] = molecule_name_atom[np.where(molecule_name_atom[:,1]==row[2])[0],3]
    add[i,3] = molecule_name_atom[np.where(molecule_name_atom[:,1]==row[3])[0],3]
    add[i,4] = molecule_name_atom[np.where(molecule_name_atom[:,1]==row[2])[0],4]
    add[i,5] = molecule_name_atom[np.where(molecule_name_atom[:,1]==row[3])[0],4]
    add[i,6] = molecule_name_atom[np.where(molecule_name_atom[:,1]==row[2])[0],5]
    add[i,7] = molecule_name_atom[np.where(molecule_name_atom[:,1]==row[3])[0],5]
    add[i,8] = molecule_name_atom[np.where(molecule_name_atom[:,1]==row[2])[0],6]
    add[i,9] = molecule_name_atom[np.where(molecule_name_atom[:,1]==row[3])[0],6]
    add[i,10] = molecule_name_atom[np.where(molecule_name_atom[:,1]==row[2])[0],7]
    add[i,11] = molecule_name_atom[np.where(molecule_name_atom[:,1]==row[3])[0],7]
    add[i,12] = molecule_name_atom[np.where(molecule_name_atom[:,1]==row[2])[0],8]
    add[i,13] = molecule_name_atom[np.where(molecule_name_atom[:,1]==row[3])[0],8]
    add[i,14] = molecule_name_atom[np.where(molecule_name_atom[:,1]==row[2])[0],9]
    add[i,15] = molecule_name_atom[np.where(molecule_name_atom[:,1]==row[3])[0],9]
    add[i,16] = molecule_name_atom[np.where(molecule_name_atom[:,1]==row[2])[0],10]
    add[i,17] = molecule_name_atom[np.where(molecule_name_atom[:,1]==row[3])[0],10]
```
